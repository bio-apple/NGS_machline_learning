````python
# Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
````

````python
#feature_selection
from sklearn.feature_selection import SelectPercentile,f_classif
selector = SelectPercentile(score_func=f_classif,percentile=95)#Percent of features to keep.
X_selected=selector.fit_transform(X_scale,y)
X_selected.shape
````

````python
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# ==============================
# PCA 计算
# ==============================
Num = 50
pca = PCA(n_components=Num)
X_pca = pca.fit_transform(X_selected)
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance[:Num])
final_value = cumulative_variance[-1]  # 前 Num 个主成分的累计方差

# ==============================
# 绘图
# ==============================
fig, axes = plt.subplots(3, 1, figsize=(16, 50), sharex=False)

# 子图 1: PCA 散点图
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette="Set2", ax=axes[0], legend='full')
axes[0].set_title('PCA Projection')
axes[0].set_xlabel('Principal Component 1')
axes[0].set_ylabel('Principal Component 2')

# 子图 2: PCA 解释方差柱状图 (百分比)
sns.barplot(x=np.arange(1, Num+1), y=explained_variance[:Num] * 100, palette='viridis', ax=axes[1])
axes[1].set_title(f'PCA: Explained Variance Ratio of Top {Num} Components')
axes[1].set_xlabel('Principal Component')
axes[1].set_ylabel('Explained Variance (%)')

# ✅ 在柱子上添加百分比
for i, val in enumerate(explained_variance[:Num] * 100):
    axes[1].text(i, val + 0.1, f"{val:.2f}%", ha='center', va='bottom', fontsize=8,rotation=45)

# 子图 3: PCA 累计解释方差曲线 (百分比)
axes[2].plot(np.arange(1, Num+1), cumulative_variance * 100, label='PCA Cumulative', color='blue')
axes[2].set_title(f'Cumulative Explained Variance (Top {Num} Components)')
axes[2].set_xlabel('Principal Component')
axes[2].set_ylabel('Cumulative Explained Variance (%)')

# ✅ 添加横虚线 (Num 的累计方差)
axes[2].axhline(y=final_value * 100, color='green', linestyle='--', linewidth=1)
axes[2].text(Num - 10, final_value * 100 + 1,
             f"Top {Num} PCs: {final_value*100:.2f}%",
             color='green', fontsize=12)

# ✅ 设置 Y 轴百分比格式
axes[1].yaxis.set_major_formatter(mticker.PercentFormatter(xmax=100))
axes[2].yaxis.set_major_formatter(mticker.PercentFormatter(xmax=100))

axes[2].legend()

# 调整布局
plt.tight_layout()
plt.show()

print(f"前 {Num} 个主成分的累计解释方差: {final_value*100:.2f}%")
````

````python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.25,random_state=42,stratify=y)
model = SVC(probability=True)  # 默认 kernel='rbf', C=1, gamma='scale'
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1] if len(y.unique()) == 2 else None

# =============================
# 4. 评价指标
# =============================
# (a) 准确率
print("准确率 (Accuracy):", accuracy_score(y_test, y_pred))

# (b) 分类报告（Precision, Recall, F1-score）
print("\n分类报告:\n", classification_report(y_test, y_pred))

# (c) 混淆矩阵
cm = confusion_matrix(y_test, y_pred)
print("\n混淆矩阵:\n", cm)

# 可视化混淆矩阵
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Default SVM)")
#10 0.9751243781094527
#20 0.9751243781094527
#50 0.9751243781094527
#100 0.9800995024875622
#150 0.9751243781094527
#200 0.9751243781094527
````
